# Act 1 — The Simplest Possible Agent

We're going to build this live, starting from a single API call. No framework yet — just us, `litellm`, and DeepSeek.

```python
import sys
sys.path.insert(0, "..")
from dotenv import load_dotenv
load_dotenv(override=True)  # reads DEEPSEEK_API_KEY from ../.env

from litellm import acompletion
```

## A reply is just one API call
No loop, no tools — ask the model something and print what comes back.

```python
response = await acompletion(
    model="deepseek/deepseek-chat",
    messages=[{"role": "user", "content": "Introduce yourself in one sentence."}],
)
print(response.choices[0].message.content)
```

## But it doesn't know what it doesn't know
Ask it something it can't reason its way to, like today's real date.

```python
response = await acompletion(
    model="deepseek/deepseek-chat",
    messages=[{"role": "user", "content": "What is today's exact date?"}],
)
print(response.choices[0].message.content)
```

It'll hedge, guess, or admit it doesn't know — it has no access to real-world state. That's the gap tools exist to fill.

## Wrap it in a function
Same call, packaged so we can reuse it.

```python
async def ask(prompt: str) -> str:
    response = await acompletion(
        model="deepseek/deepseek-chat",
        messages=[{"role": "user", "content": prompt}],
    )
    return response.choices[0].message.content

print(await ask("What is 2 + 2?"))
```

That's the whole first version of our agent: a function that replies. It has no memory (every call is independent), no tools, no loop.
