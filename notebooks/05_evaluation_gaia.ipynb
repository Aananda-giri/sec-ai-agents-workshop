# Act 5 — Evaluating on GAIA

GAIA (General AI Assistants benchmark) is a set of real-world questions requiring reasoning, tool use, and multi-step problem solving. The real dataset is gated on HuggingFace; we're using a handful of GAIA-style questions bundled locally so this runs with zero setup.

```python
import sys
sys.path.insert(0, "..")
from dotenv import load_dotenv
load_dotenv(override=True)

from mini_agent import Agent, calculator, execute_python, web_search
from mini_agent.gaia import SAMPLE_GAIA, evaluate
```

## The questions

```python
for q in SAMPLE_GAIA:
    print("-", q["question"])
```

## Baseline: an agent with no tools
Using 2 questions live to respect the free-tier rate limit (see README). Full offline run: `evaluate(make_agent, SAMPLE_GAIA)`.

```python
def make_baseline_agent():
    return Agent(instructions="Answer with just the final answer, no explanation.", max_steps=3)

baseline_results = await evaluate(make_baseline_agent, problems=SAMPLE_GAIA[:2])
```

## Now give it tools

```python
def make_tool_agent():
    return Agent(
        tools=[calculator, execute_python],
        instructions="Answer with just the final answer, no explanation. Use tools for anything computational.",
        max_steps=5,
    )

tool_results = await evaluate(make_tool_agent, problems=SAMPLE_GAIA[:2])
```

## One from the real benchmark
Everything above is arithmetic — the real GAIA benchmark usually needs actual research. This is one of its most-cited real questions: the agent has to look up two live facts (Kipchoge's marathon pace, the Moon's minimum perigee) and combine them. Guessing from training data isn't enough.

```python
kipchoge = SAMPLE_GAIA[-1]
print(kipchoge["question"])
print("expected:", kipchoge["answer"])
```

No tools — it has to guess:

```python
agent = Agent(instructions="Answer with just the final answer, no explanation.", max_steps=3)
answer = await agent.run(kipchoge["question"])
print("\nFINAL:", answer)
```

With `web_search` back in the toolbox from Act 2 — now it can go look the facts up instead of guessing:

```python
agent2 = Agent(
    tools=[calculator, execute_python, web_search],
    instructions=(
        "Answer with just the final answer, no explanation. "
        "Use web_search at most twice (once for the marathon pace, once for the moon perigee), "
        "then use execute_python to do the final calculation precisely."
    ),
    max_steps=8,
)
answer2 = await agent2.run(kipchoge["question"])
print("\nFINAL:", answer2)
```

Right facts, right arithmetic — and it may *still* score as wrong. The question asks for the count in **thousands** of hours (`17`); it's easy for the model to instead report the raw hour count rounded to the nearest thousand (`17000`). Same brittleness as the train-speed question earlier, at real GAIA's actual difficulty.
